# We know that, we are presuming that our sampled data of 569 observations of radius mean is being sampled from Rayleigh Distributed population having some unknwon population mode parameter. 

# And now from Maximum Likelihood Estimation, we know that the good point estimator (MVUE) of the population parameter (population mode) of rayleigh distributed population is given by: 

\begin{equation}
\hat{\theta_1} = \sqrt{\frac{\sum_{i=0}^{568}x_i^2}{2.569}}
\end{equation}

# We are now going to calculate the value of theta 1 hat for the sample of 569 oservations on the assumption that these observations are being sampled from a rayleigh distributed population. The value of theta 1 hat (MVUE) for the sample of 569 observations is going to give us the best point estimate of the unknown population parameter of rayleigh distributed population. 

```python
X
```

```python
theta_1_hat = np.sqrt(np.mean(X**2)/2)
```

```python
theta_1_hat
```

```python
log_like_max_value = np.sum(s.rayleigh.logpdf(X,scale=theta_1_hat))
```

```python
log_like_max_value
```

```python
def compute_log_likelihood_value(theta_1_hat):
    
    return np.sum(s.rayleigh.logpdf(X,scale=theta_1_hat))
```

```python
theta_1_hat = np.arange(9,12,0.01)
```

```python
theta_1_hat.shape
```

```python
log_e_L = []

for theta_1 in theta_1_hat:
    
    log_e_L.append(compute_log_likelihood_value(theta_1))
```

```python
log_e_L
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(theta_1_hat,log_e_L)

plt.xlabel("Different Values of Good Point Estimator of Population Paramter of Rayleigh Distribution")

plt.ylabel("Different Values of Log Likelihood Function for Rayleigh Distribution")
```

```python
max(log_e_L)
```
